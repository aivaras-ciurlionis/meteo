```python
from PIL import ImageChops, Image
from matplotlib.pyplot import imshow
from matplotlib import pyplot
import numpy as np
```

```python
def show(img):
    np.asarray(img)
    imshow(np.asarray(img))

def show2(im1, im2):
    f, axis = pyplot.subplots(1,2)
    axis[0].imshow(np.asarray(im1))
    axis[1].imshow(np.asarray(im2))
    
def show_all(images):
    f, axis = pyplot.subplots(1,len(images), figsize=(15, 6))
    for i, im in enumerate(images):
        axis[i].imshow(np.asarray(im))
```

```python
img1 = Image.open('../pics/2017-10-23--09-00-00.png')
img2 = Image.open('../pics/2017-10-23--09-15-00.png')
img3 = Image.open('../pics/2017-10-23--09-30-00.png')
img4 = Image.open('../pics/2017-10-23--10-45-00.png')
show_all([img1, img2, img3, img4])
```

```python
diff1 = ImageChops.difference(img1,img2)
diff2 = ImageChops.difference(img2,img3)
diff3 = ImageChops.difference(img3,img4)

dif_max= ImageChops.difference(img4,img1)
show_all([diff1,diff2,diff3, dif_max])
```
